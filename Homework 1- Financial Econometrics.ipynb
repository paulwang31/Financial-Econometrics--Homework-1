## The data from 1957-01-01 to today

```r
# Make defination to compute mean, variance, skewness and kurtosis coefficient for each stock.
adjustprice <- function(st, putset){df = getSymbols(st, src = "yahoo",auto.assign = FALSE, from ="1957-01-01")
df=na.omit(df)
logdf=log(df[,6])
logdf=na.omit(logdf)
rtndf=diff(logdf[,1])*100
rtndf[1,1]=0
T = length(rtndf[,1])
mu = 1/(T) * sum(rtndf[,1])
sig2 = 1/(T-1) * sum((rtndf[,1]-mu)^2)
sk = 1/((T-1) * sqrt(sig2)^3) * sum((rtndf[,1]-mu)^3)
ku = 1/((T-1) * sqrt(sig2)^4) * sum((rtndf[,1]-mu)^4)-3
putsubset=data.frame(row.names = st)
putsubset[1,1]=mu
putsubset[1,2]=sig2
putsubset[1,3]=sk
putsubset[1,4]=ku
names(putsubset)=c('Mean','Variance','Skewness','Kurtosis')
putset<-rbind(putset,putsubset)
return(putset)
}
```

```r
#Load the tickers 
tickers=read.csv('/Users/wangkewei/Desktop/Econometrics/Homework1/tickers2.csv')
```

```r
library("quantmod")
#Produce the matrix
df=getSymbols('^GSPC', src = "yahoo",auto.assign = FALSE, from ="1957-01-01")
df=na.omit(df)
logdf=log(df[,6])
logdf=na.omit(logdf)
rtndf=diff(logdf[,1])*100
rtndf[1,1]=0
T = length(rtndf[,1])
mu = 1/(T) * sum(rtndf[,1])
sig2 = 1/(T-1) * sum((rtndf[,1]-mu)^2)
sk = 1/((T-1) * sqrt(sig2)^3) * sum((rtndf[,1]-mu)^3)
ku = 1/((T-1) * sqrt(sig2)^4) * sum((rtndf[,1]-mu)^4)-3
putset=data.frame(row.names = 'S&P500')
putset[1,1]=mu
putset[1,2]=sig2
putset[1,3]=sk
putset[1,4]=ku
names(putset)=c('Mean','Variance','Skewness','Kurtosis')

for (st in as.character(tickers$X0)){
  putset=adjustprice(st, putset)
  adjustprice(st, putset)
}
```

### Question 1&5

#### Mean

```r
#Mean
library(muStat)
d <- density(putset$Mean)
dn <- dnorm(d$x, mean(putset$Mean), stdev(putset$Mean, unbiased=FALSE))
plot(d$x,d$y,type='l', col="red", main=c("Mean", "empirical density vs normal"), xlab="Mean", ylab="Density")
lines(d$x,dn)
```

```r
#Kolmogorov Smirnov test
ks.test(putset$Mean,'pnorm')
```

>For the total dataset, we can see that the distribution of the mean part has zero skewness, as its mean is almost the same with the normal one, thus we can say the distribution of the mean part is symmetrical. However, the kurtosis is much more than the normal one, which means the mean part's data is more distributed dispersively than the normal one.<br>
For the ks test, we can see that this part is obviously not following the normal distribution and thus it is not stable.
